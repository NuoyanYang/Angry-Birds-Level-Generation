```python
import numpy as np
import matplotlib.pyplot as plt
```

*   Define fitness function: 

```python
def fitness(x):
    return np.sin(x)
```

*   Define individuals:

```python
upper = 20
lower = -20
pop_size  =15
population = np.random.uniform(lower, upper, pop_size)
```

```python
population
```

*   Plot function:

```python
domain = np.arange(lower, upper, 0.001)
plt.plot(domain, fitness(domain), 'b')

plt.title("Some title")
plt.xlim([lower,upper])
plt.ylim([-6,6])
plt.show()
```

*   Plot function and individuals:

```python
domain = np.arange(lower, upper, 0.001)
plt.plot(domain, fitness(domain), 'b')

plt.scatter(population, fitness(population), c = 'r')

plt.title("Some title")
plt.xlim([lower,upper])
plt.ylim([-6,6])
plt.show()
```
